```python
def has_path(graph, src, dst):
  queue = []
  queue.append(src)
  idx = 0
  while idx < len(queue):
    current = queue[idx]
    idx += 1
    
    for node in graph[current]:
      if node == dst:
        return True
      queue.append(node)
  return False


graph = {
  'f': ['g', 'i'],
  'g': ['h'],
  'h': [],
  'i': ['g', 'k'],
  'j': ['i'],
  'k': []
}

print(has_path(graph, 'f', 'k')) # True
```

```python
def has_path(graph, src, dst):
  if(src == dst): 
    return True
  for node in graph[src]:
    if has_path(graph, node, dst) == True:
      return True
  return False


graph = {
  'f': ['g', 'i'],
  'g': ['h'],
  'h': [],
  'i': ['g', 'k'],
  'j': ['i'],
  'k': []
}

print(has_path(graph, 'f', 'k')) # True
```

```python
def undirected_path(edges, node_A, node_B):
  graph = Graph(edges)
  stack = [node_A]
  visited = []
  while len(stack) > 0:
    current = stack.pop()
    visited.append(current)
    
    for node in graph.data[current]:
      if node == node_B:
        return True
      if not node in visited:
        stack.append(node)
  return False
    
class Graph:
  def __init__(self, edges):
    self.edges = edges
    self.data = {}
    for edge in edges:
      n1, n2 = edge
      if n1 not in self.data:
        self.data[n1] = []
      if n2 not in self.data:
        self.data[n2] = []
        
      self.data[n1].append(n2)
      self.data[n2].append(n1)

        
        
edges = [
  ('i', 'j'),
  ('k', 'i'),
  ('m', 'k'),
  ('k', 'l'),
  ('o', 'n')
]

print(undirected_path(edges, 'j', 'm') )# -> True
```

```python
def connected_components_count(graph):
  count = 0
  visited = []
  for node in graph:
    if traversal(graph, node, visited) == True:
      count += 1
  return count

def traversal(graph, src, visited):
  if src in visited:
    return False
  
  visited.append(src)
  
  for neighbour in graph[src]:
    traversal(graph, neighbour, visited)
    
  return True

print(connected_components_count({
  0: [8, 1, 5],
  1: [0],
  5: [0, 8],
  8: [0, 5],
  2: [3, 4],
  3: [2, 4],
  4: [3, 2]
})) # -> 2
      
```

```python
def largest_component(graph):
  visited = []
  largest = 0
  for node in graph:
    result = traverse(graph, node, visited)
    if result > largest:
      largest = result
  return largest
      
    
def traverse(graph, src, visited):
  if src in visited:
    return 0
  size = 1
  visited.append(src)
  for neighbour in graph[src]:
    size += traverse(graph, neighbour, visited)
    
  return size
             
             
print(largest_component({
  0: [8, 1, 5],
  1: [0],
  5: [0, 8],
  8: [0, 5],
  2: [3, 4],
  3: [2, 4],
  4: [3, 2]
})) # -> 4
```

```python
def shortest_path(edges, node_A, node_B):
  graph = Graph(edges)
  visited = [node_A]
  queue = [(node_A, 0)]
  idx = 0
  
  while idx < len(queue):
    current_node, current_weight = queue[idx]
    if current_node == node_B:
      return current_weight
    idx += 1
    
    for node in graph.data[current_node]:
      if not node in visited:
        visited.append(current_node)
        queue.append((node, current_weight + 1))
  return -1    

class Graph:
  def __init__(self, edges):
    self.data = {}
    for edge in edges:
      n1, n2 = edge
      if not n1 in self.data:
        self.data[n1] = []
      if not n2 in self.data:
        self.data[n2] = []
      self.data[n1].append(n2)
      self.data[n2].append(n1)
edges = [
  ['w', 'x'],
  ['x', 'y'],
  ['z', 'y'],
  ['z', 'v'],
  ['w', 'v']
]

print(shortest_path(edges, 'w', 'z')) # -> 2
```

```python
def island_count(grid):
  visited = []
  result = 0
  for r in range(len(grid)):
    for c in range(len(grid[0])):
      if traverse(grid, r, c, visited) == True:
        result += 1
  return result
      

def traverse(grid, r, c, visited):
  row_in_bound = 0 <= r and r < len(grid)
  column_in_bound = 0 <= c and c < len(grid[0])
  
  if row_in_bound == False or column_in_bound == False:
    return False
  
  if grid[r][c] == 'W':
    return False
  
  pos = str(r) + ',' + str(c)
  if pos in visited:
    return False
  
  visited.append(pos)
  
  traverse(grid, r, c - 1, visited) #up
  traverse(grid, r + 1, c, visited) #right
  traverse(grid, r, c + 1, visited) #down
  traverse(grid, r - 1, c, visited) #left
  
  return True
  
 



grid = [
  ['W', 'L', 'W', 'W', 'W'],
  ['W', 'L', 'W', 'W', 'W'],
  ['W', 'W', 'W', 'L', 'W'],
  ['W', 'W', 'L', 'L', 'W'],
  ['L', 'W', 'W', 'L', 'L'],
  ['L', 'L', 'W', 'W', 'W'],
]

print(island_count(grid)) # -> 3
```

```python
def minimum_island(grid):
  visited = []
  min_isl = float('inf')
  for r in range(len(grid)):
    for c in range(len(grid[0])):
      count = traverse(grid, r, c, visited)
      if count > 0 and count < min_isl:
        min_isl = count
  return min_isl
        
        
      
def traverse(grid, r, c, visited):
  row_in_bound = 0 <= r and r < len(grid)
  column_in_bound = 0 <= c and c < len(grid[0])
  
  if row_in_bound == False or column_in_bound == False:
    return 0
  
  if grid[r][c] == 'W':
    return 0
  
  pos = str(r) + ',' + str(c)
  
  if pos in visited:
    return 0
  
  visited.append(pos)

  count = 1
  count += traverse(grid, r, c-1, visited) # up
  count += traverse(grid, r+1, c, visited) # right
  count += traverse(grid, r, c+1, visited) # down
  count += traverse(grid, r-1, c, visited) # left
  
  return count

grid = [
  ['W', 'L', 'W', 'W', 'W'],
  ['W', 'L', 'W', 'W', 'W'],
  ['W', 'W', 'W', 'L', 'W'],
  ['W', 'W', 'L', 'L', 'W'],
  ['L', 'W', 'W', 'L', 'L'],
  ['L', 'L', 'W', 'W', 'W'],
]

print(minimum_island(grid)) # -> 2
```
